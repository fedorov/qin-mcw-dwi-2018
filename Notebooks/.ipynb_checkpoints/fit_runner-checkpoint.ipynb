```python
import os, sys, subprocess

dataPath = '/Users/fedorov/Dropbox (Partners HealthCare)/ImageData/QIN/2018-QIN-DWI_challenge/PatientData-mpReview'
analysisPath = '/Users/fedorov/Dropbox (Partners HealthCare)/ImageData/QIN/2018-QIN-DWI_challenge/PatientData-maps'

knownModels = ['MonoExponential','BiExponential','Kurtosis','Gamma','StretchedExponential']


def fitModel(inputImage,modelName, outputDir,prefix=''):
    #commonParameters = ['fittedVolume','rsqrVolume',
    if not modelName in knownModels:
        print("ERROR: Unknown model")
        return
    
    parameters = {}
    parameters['MonoExponential'] = ['adcMonoExpDiff']
    parameters['BiExponential'] = ['slowDiff','fastDiff','fastDiffFraction']
    parameters['Kurtosis'] = ['kurtosisDiff','kurtosis']
    parameters['Gamma'] = ['theta','k','mode']
    parameters['StretchedExponential'] = ['ddc','alpha']
    
    cli = '/Applications/Slicer.app/Contents/Extensions-27269/SlicerProstate/lib/Slicer-4.9/cli-modules/DWModeling'
    cmd = [cli,'--model',modelName]
    for p in parameters[modelName]:
        if prefix:
            fileName = prefix+'_'+modelName+'_'+p+'.nrrd'
        else:
            fileName = modelName+'_'+p+'.nrrd'
        cmd = cmd+['--'+p,os.path.join(outputDir,fileName)]
        
    cmd = cmd+[inputImage]
    subprocess.run(cmd)
    
runCnt = 0
for root, dirs, files in os.walk(dataPath):
    for file in files:
        if not file.endswith('.nrrd'):
            continue
        rootPath = root.split('/')
        if not int(rootPath[-2])>10:
            continue
        # have a DWI reconstruction
        patientID = rootPath[-4]
        series = rootPath[-2]
        print('Subject '+patientID+' - Series '+series)
        
        for m in knownModels:
            print(' '+m)
            thisAnalysisPath = os.path.join(analysisPath,patientID[:-2],series,m)
            thisAnalysisPath = analysisPath
            prefix=patientID[:-2]+'-'+series
            os.makedirs(thisAnalysisPath,exist_ok=True)
            fitModel(inputImage=os.path.join(root,file),modelName=m,outputDir=thisAnalysisPath,prefix=prefix)
            runCnt = runCnt+1
        
```
